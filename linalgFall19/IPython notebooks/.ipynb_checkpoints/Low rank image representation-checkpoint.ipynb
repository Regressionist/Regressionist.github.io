```python
import numpy as np
from numpy.linalg import svd
import cv2
import matplotlib.pyplot as plt
```

```python
img_path = 'Data/Cityscapes/leftImg8bit/train/zurich/zurich_000002_000019_leftImg8bit.png'
img = cv2.imread(img_path)
imggray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(imggray,cmap='gray')
plt.show()
```

```python
def compress_img(matrix,k,img_title):
    U,s,V = svd(matrix, full_matrices=False)
    plt.imshow(np.dot(U[:,0:k],np.dot(np.diag(s[:k]),V[0:k,:])),cmap='gray')
    plt.title(img_title)
    plt.show()
```

```python
for rank in [10,50,100,200,500]:
    compress_img(imggray,rank,'rank {}'.format(rank))
```
