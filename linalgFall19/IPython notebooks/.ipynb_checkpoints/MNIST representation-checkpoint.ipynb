```python
from mnist import MNIST
import numpy as np
import matplotlib.pyplot as plt
```

```python
mndata = MNIST('data/MNIST/')
images, labels = mndata.load_training()
labels = np.array(labels)
```

```python
unique_number_idx = [np.where(labels==i)[0][0] for i in range(0,10)]
images_subset = [np.reshape(images[idx],(28,28)) for idx in unique_number_idx]
```

```python
fig = plt.figure(figsize=(18,18))
for i in range(0,10):
    ax = fig.add_subplot(1,10,i+1)
    ax.imshow(images_subset[i], cmap='gray')
    ax.axis('off')
```

```python
images_matrix = np.array([images[idx] for idx in range(10000)])
mu = np.mean(images_matrix, axis=0)
data = (images_matrix - mu).T
```

```python
U,S,Vt = np.linalg.svd(data)
```

```python
plt.imshow(np.reshape(U[:,0],(28,28)), cmap='gray')
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize = (10,4))
reduced_1d_data = np.matmul(data.T, U[:,0])
l = labels[:5000]
num = [0,1]
#markers=['o','^']
for i in num:
    ind = np.where(l==i)[0]
    plt.scatter(reduced_1d_data[ind], len(ind)*[0], s=50, label='{}'.format(i))

plt.legend()
plt.axis('off')
plt.show()
```

```python
fig = plt.figure(figsize=(125,100))
for i in range(0,60):
    ax = fig.add_subplot(6,10,i+1)
    ax.imshow(np.reshape(U[:,i],(28,28)), cmap='gray')
    ax.axis('off')
```
