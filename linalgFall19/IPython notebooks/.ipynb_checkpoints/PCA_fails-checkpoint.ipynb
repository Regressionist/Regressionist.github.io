```python
from mnist import MNIST
import numpy as np
import matplotlib.pyplot as plt
```

```python
mndata = MNIST('data/MNIST/')
images, labels = mndata.load_training()
```

```python
unique_number_idx = [labels.index(i) for i in range(0,10)]
images_subset = [np.reshape(images[idx],(28,28)) for idx in unique_number_idx]
```

```python
fig = plt.figure(figsize=(18,18))
for i in range(0,10):
    ax = fig.add_subplot(1,10,i+1)
    ax.imshow(images_subset[i], cmap='gray')
    ax.axis('off')
```

```python
def add_gaussian_noise(img, mean, std_dev):
    noise = np.random.normal(loc = mean, scale = std_dev, size = (img.shape[0], img.shape[1]))
    return img+noise
```

```python
mean = 0
std_dev = 10
noisy_imgs = [add_gaussian_noise(img, mean, std_dev) for img in images_subset]
```

```python
fig1 = plt.figure(figsize=(18,18))
fig2 = plt.figure(figsize=(18,18))
for i in range(0,10):
    ax1 = fig1.add_subplot(1,10,i+1)
    ax1.imshow(images_subset[i], cmap='gray')
    ax1.axis('off')
    ax2 = fig2.add_subplot(1,10,i+1)
    ax2.imshow(noisy_imgs[i], cmap='gray')
    ax2.axis('off')
```

```python
covariance = np.cov(noisy_imgs[1].T)
v,w = np.linalg.eig(covariance)
```

```python
fig = plt.figure(figsize=(8,8))
plt.bar(range(0,len(v)), sorted(v, reverse=True), width=0.5)
plt.show()
```

## Implementation 

```python
def denoise(noisy_image, k):
    U,sig,Vt = np.linalg.svd(noisy_image)
    return np.matrix(U[:, :k]) * np.diag(sig[:k]) * np.matrix(Vt[:k, :])
```

```python
denoised_imgs = [denoise(img, 6) for img in noisy_imgs]
```

```python
fig1 = plt.figure(figsize=(18,18))
fig2 = plt.figure(figsize=(18,18))
fig3 = plt.figure(figsize=(18,18))
for i in range(0,10):
    ax1 = fig1.add_subplot(1,10,i+1)
    ax1.imshow(images_subset[i], cmap='gray')
    ax1.axis('off')
    ax2 = fig2.add_subplot(1,10,i+1)
    ax2.imshow(noisy_imgs[i], cmap='gray')
    ax2.axis('off')
    ax3 = fig3.add_subplot(1,10,i+1)
    ax3.imshow(denoised_imgs[i], cmap='gray')
    ax3.axis('off')
```
